```python
birth_gender = user_spec.groupby(['user_id','birth_year','gender'], as_index=False).count().reset_index(drop=True).iloc[:,:3]

birth_gender
```

```python
user_spec = user_spec.drop(['birth_year','gender'], axis=1)

user_spec = pd.merge(user_spec, birth_gender, how='left', on='user_id')

user_spec
```

```python
user_spec = user_spec.replace('생활비', 'LIVING')
user_spec = user_spec.replace('대환대출', 'SWITCHLOAN')
user_spec = user_spec.replace('투자', 'INVEST')
user_spec = user_spec.replace('기타', 'ETC')
user_spec = user_spec.replace('주택구입', 'BUYHOUSE')
user_spec = user_spec.replace('전월세보증금', 'HOUSEDEPOSIT')
user_spec = user_spec.replace('자동차구입', 'BUYCAR')
user_spec = user_spec.replace('사업자금', 'BUSINESS')
```

```python
user_spec['purpose'].unique()
```

# ex loan cnt & ex loan amt

```python
# existing loan_count = na -> 0

user_spec['existing_loan_cnt'].fillna(0, inplace=True)
```

```python
user_spec.isnull().sum()
```

```python
# if existing loan count = 0, existing loan amount = 0

user_spec.loc[user_spec['existing_loan_cnt'] == 0, 'existing_loan_amt'] = 0
```

```python
user_spec.isnull().sum()
```

# company enter month -> yyyy.mm

```python
# yyyy.mm
user_spec[user_spec['company_enter_month'] < 1000000]
```

```python
# yyyy.mm.dd
user_spec[user_spec['company_enter_month'] > 1000000]
```

```python
user_spec['company_enter_month'] = user_spec.apply(lambda x: x.company_enter_month // 100 if x.company_enter_month > 10000000 else x.company_enter_month, axis=1)

user_spec
```

```python
# yyyy.mm.dd
user_spec[user_spec['company_enter_month'] > 1000000]
```

# company_enter < birth

```python
user_spec[user_spec['birth_year'] > (user_spec['company_enter_month'] // 100)]
```

```python
user_spec['company_year'] = user_spec['company_enter_month'] // 100
user_spec['company_month'] = user_spec['company_enter_month'] % 100

user_spec
```

```python
# company enter before birth

user_spec.loc[user_spec['birth_year'] > user_spec['company_year'], 'company_enter_month'] = None

user_spec
```

```python
user_spec[user_spec['birth_year'] > user_spec['company_year']]
```

```python
user_spec = user_spec.drop(['company_year','company_month'], axis=1)

user_spec
```

# merge user_spec, loan_result on app_id

```python
user_loan = pd.merge(user_spec, loan_result, 'left', on='application_id')

user_loan
```

```python
user_loan.to_csv(os.path.join(data_dir, 'user_loan.csv'))
```

# merge user_loan, log_data on user_id

```python
main_df = pd.merge(user_loan, log_data, 'left', on='user_id')

main_df
```

```python
main_df.to_csv(os.path.join(data_dir, 'main_df.csv'))
```

```python
# 1) training set에서 표본 크기가 n인 bootstrap sampling 수행
# - bootstrap sampling: original sample 집단에서 더 작지만 무수히 많은 집단으로 랜덤하게 뽑는 방법

# 2) bootstrap sample에 대해 random forest tree 모형 제작
# - 전체 변수 중 m개 변수를 랜덤하게 선택
# - 최적의 classifier 선정
# - classifier에 따라 두 개의 daughter node 생성

# 3) tree들의 앙상블 학습 결과 출력
# - ensembl learning: 큰 데이터를 수많은 작은 set으로 나눠 학습시키고, 각 학습 모델을 연결해 성능 좋은 머신러닝 구축
```
